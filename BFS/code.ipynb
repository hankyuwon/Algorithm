## 알고리즘 수업 - 너비 우선 탐색 1
### 실버 2
### https://www.acmicpc.net/problem/24444

```python
from collections import deque
import sys
input = sys.stdin.readline
# N: 정점의 수 , M: 간선의 수, R: 시작 정점
N, M, R = map(int, input().split())

graph = [[] for _ in range(N+1)]

# BFS 탐색

for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

visited = [0]*(N+1)
visited[R] = 1

queue = deque([R])

cnt = 1

while queue:
    node = queue.popleft()
    graph[node].sort()
    
    for i in graph[node]:
        if not visited[i]:
            queue.append(i)
            cnt += 1
            visited[i] = cnt
            
for i in visited[1:]:
    print(i)
```

### 알고리즘 수업 - 너비 우선 탐색 2

```python
from collections import deque
import sys
input = sys.stdin.readline

# N: 정점의 수 , M: 간선의 수, R: 시작 정점
N, M, R = map(int, input().split())

graph = [[] for _ in range(N+1)]

# BFS 탐색

for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

visited = [0]*(N+1)
visited[R] = 1

queue = deque([R])

cnt = 1

while queue:
    node = queue.popleft()
    graph[node].sort(reverse=True)
    
    for i in graph[node]:
        if not visited[i]:
            queue.append(i)
            cnt += 1
            visited[i] = cnt
            
for i in visited[1:]:
    print(i)
```

## 미로 탐색
### 실버 1
### https://www.acmicpc.net/problem/2178

```python
import sys
from collections import deque

input = sys.stdin.readline

n, m = map(int, input().split())
graph = []

for _ in range(n):
    graph.append(list(map(int, input().rstrip()))) # readline의 경우 맨 뒤에 '\n'까지 입력받으므로 제거해줘야 함

# 상하좌우
dx = [-1, 1, 0, 0] 
dy = [0, 0, -1, 1]

def bfs(x, y):
    
    queue = deque()
    queue.append((x,y))

    while queue:
        x, y = queue.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m and graph[nx][ny] == 1:
                queue.append((nx, ny))
                graph[nx][ny] = graph[x][y] + 1
    
    return graph[n-1][m-1]

print(bfs(0,0))
```

## 나이트의 이동
### 실버 1
### https://www.acmicpc.net/problem/7562

```python
from collections import deque

n = int(input())

dx = [1, 2, 2, 1, -1, -2, -2, -1]
dy = [2, 1, -1, -2, -2, -1, 1, 2]


def bfs(x0, y0, l, x1, y1):
    
    # graph = [[0]*l]*l
    graph = [[0]*l for _ in range(l)]
    
    queue = deque()
    queue.append((x0,y0))
    graph[x0][y0] = 1
    
    while queue:
        x, y = queue.popleft()
        
        if x == x1 and y == y1:
            return graph[x][y] - 1
        
        for i in range(8):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < l and 0 <= ny < l:
                if graph[nx][ny] == 0:
                    queue.append((nx,ny))
                    graph[nx][ny] = graph[x][y] + 1
        
for _ in range(n):
    l = int(input()) # 체스판 한 변의 길이
    x0, y0 = list(map(int, input().split())) # 시작 점 x0, y0 입력
    x1, y1 = list(map(int, input().split())) # 도착 점 x1, y1 입력

    ans = bfs(x0,y0,l,x1,y1)
    print(ans)
```

## 트리의 부모 찾기
### 실버 2
### https://www.acmicpc.net/problem/11725

```python
import sys
from collections import deque

input = sys.stdin.readline
n = int(input())

tree = [[] for _ in range(n+1)]

# print(tree)

for _ in range(n-1):
    a, b = list(map(int, input().split()))
    tree[a].append(b)
    tree[b].append(a)
    
# print(tree)

def dfs(graph, start):
    stack = [start]
    visited = [-1] * (n+1)
    result = [0] * (n+1)
    
    while stack:
        cnt_node = stack.pop()
        if visited[cnt_node] == 1:
            continue
        visited[cnt_node] = 1
        
        for adj_node in graph[cnt_node]:
            if visited[adj_node] == -1:
                stack.append(adj_node)
                result[adj_node] = cnt_node
    print(*result[2:], sep = '\n')

dfs(tree,1)
```

## 토마토
### 골드 5
### https://www.acmicpc.net/problem/7576

```python
import sys
from collections import deque

input = sys.stdin.readline
a, b = list(map(int, input().split()))

graph = []

for _ in range(b):
    graph.append(list(map(int, input().split())))

visited = [[0]*a for _ in range(b)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(_graph, _queue):
    
    max = -1
    
    while _queue:
        x, y = _queue.popleft()
        visited[x][y] = 1
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < b and 0 <= ny < a and visited[nx][ny] == 0 and _graph[nx][ny] != -1:
                _queue.append((nx,ny))
                visited[nx][ny] = 1
                _graph[nx][ny] = _graph[x][y] + 1
                if _graph[nx][ny] >= max:
                    max = _graph[nx][ny]
    
    for i in range(b):
        for j in range(a):
            if _graph[i][j] == 0:
                return -1
    
    if max <= 1:
        return 0
    else:
        return int(max-1)

queue = deque()

for i in range(b):
    for j in range(a):
        if graph[i][j] == 1:
            queue.append((i,j))
            visited[i][j] = 1

c = bfs(graph, queue)
print(c)
```

## 토마토(2)
### 골드 5
### https://www.acmicpc.net/problem/7569

```python
import sys
from collections import deque

input = sys.stdin.readline

a, b, c = list(map(int, input().split()))


graph = []
for _ in range(c):
    temp = []
    for _ in range(b):
        temp.append(list(map(int, input().split())))
    graph.append(temp)


visited = [[[0]*a for _ in range(b)] for _ in range(c)]


dx = [1,0,-1,0,0,0]
dy = [0,1,0,-1,0,0]
dz = [0,0,0,0,1,-1]

def bfs(graph, queue):
    max = -1
    
    while queue:
        z, y, x = queue.popleft()
        visited[z][y][x] = 1
        
        for i in range(6):
            nz = z + dz[i]
            ny = y + dy[i]
            nx = x + dx[i]
        
            if 0 <= nx < a and 0 <= ny < b and 0 <= nz < c and graph[nz][ny][nx] != -1 and visited[nz][ny][nx] == 0:
                queue.append((nz,ny,nx))
                visited[nz][ny][nx] = 1
                graph[nz][ny][nx] = graph[z][y][x] + 1
                if graph[nz][ny][nx] >= max:
                    max = graph[nz][ny][nx]

    for i in range(c):
        for j in range(b):
            for k in range(a):
                if graph[i][j][k] == 0:
                    return -1
    if max <= 1:
        return 0
    else:
        return int(max-1)

queue = deque()

for i in range(c):
    for j in range(b):
        for k in range(a):
            if graph[i][j][k] == 1:
                queue.append((i,j,k))
                visited[i][j][k] = 1

response = bfs(graph, queue)
print(response)
```
