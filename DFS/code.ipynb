## 알고리즘 수업 - 깊이 우선 탐색 1
### 실버 2
### https://www.acmicpc.net/problem/24479

```python
import sys
input = sys.stdin.readline

N, M, R = map(int,input().split())

graph = [[] for _ in range(N+1)]
for _ in range(M):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
for i in range(1, len(graph)):
    graph[i].sort(reverse=True)

def DFS(start):
    stack = [start]
    visited = [-1]*(N+1)
    result = [0]*(N+1)
    cnt = 1
    
    while stack:
        cnt_node = stack.pop()
        if visited[cnt_node] == 1:
            continue
        visited[cnt_node] = 1
        
        result[cnt_node] = cnt
        cnt += 1
        
        for adj_node in graph[cnt_node]:
            if visited[adj_node] == -1:
                stack.append(adj_node)
    return result

result = DFS(R)

print(*result[1:], sep='\n')
```

## DFS와 BFS
### 실버 2
### https://www.acmicpc.net/problem/1260

```python
from collections import deque
import sys
input = sys.stdin.readline

N, M, R = map(int, input().split())

graph = [[] for _ in range(N+1)]

for _ in range(M):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

#DFS
def DFS(R):
    _graph = graph.copy()
    
    for i in range(1, len(graph)):
        _graph[i].sort(reverse= True)
        
    stack = [R]
    visited = [-1]*(N+1)
    result = [0]*(N+1)
    cnt = 1
    
    while stack:
        cnt_node = stack.pop()
        if visited[cnt_node] == 1:
            continue
        print(cnt_node, end = ' ')
        visited[cnt_node] = 1
        
        result[cnt_node] = cnt
        cnt += 1
        
        for adj_node in graph[cnt_node]:
            if visited[adj_node] == -1:
                stack.append(adj_node)                


#BFS
def BFS(R):
    visited = [0]*(N+1)
    visited[R] = 1
    
    queue = deque([R])
    cnt = 1
    
    while queue:
        node = queue.popleft()
        print(node, end = ' ')
        graph[node].sort()
        
        for i in graph[node]:
            if not visited[i]:
                queue.append(i)
                cnt += 1
                visited[i] = cnt


DFS(R)
print()
BFS(R)
```

## 바이러스
### 실버 3
### https://www.acmicpc.net/problem/2606

```python
N = int(input())
V = int(input())

graph = [[] for _ in range(N+1)]

for _ in range(V):
    a,b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
def DFS(start):
    stack = [start]
    visited = [-1]*(N+1)
    result = [0]*(N+1)
    
    while stack:
        temp_node = stack.pop()
        if visited[temp_node] == 1:
            continue
        
        visited[temp_node] = 1
        
        result[temp_node] = 1
        
        for adj_node in graph[temp_node]:
            if visited[adj_node] == -1:
                stack.append(adj_node)
    return result
result = DFS(1)

print(result.count(1)-1)
```

## 안전 영역
### 실버 1
### https://www.acmicpc.net/problem/2468

```python
import sys
sys.setrecursionlimit(100000)

n = int(input())
graph = []
max_num = 0
result = 1

dx = [0,0,-1,1]
dy = [1,-1,0,0]

for i in range(n):
    low = list(map(int, input().split()))
    graph.append(low)

    if max(low) > max_num:
        max_num = max(low)


def dfs(x,y,num):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]

        if 0<=nx<n and 0<=ny<n and visited[nx][ny] == 0:
            if graph[nx][ny] > num:
                visited[nx][ny] = 1
                dfs(nx,ny,num)

for i in range(max_num):
    visited = [[0]*n for _ in range(n)]
    cnt = 0

    for j in range(n):
        for k in range(n):
            if graph[j][k] > i and visited[j][k] == 0:
                cnt += 1
                visited[j][k] = 1
                dfs(j,k,i)
    result = max(result, cnt)

print(result)
```
